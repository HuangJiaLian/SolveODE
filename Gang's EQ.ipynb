```python
import numpy as np
import matplotlib.pyplot as plt 

def f(q):
    return 0.5*q**3

def df2(q):
    return 3*q

def df1(q):
    return 1.5*q**2

# # fixed 3-4 model
# def f(q):
#     """For 3-4-p sphere model."""
#     return 0.5*(q**3 + q**4)
# def df2(q):
#     """For 3-4-p sphere model."""
#     return 3*q + 6*q**2
# def df1(q):
#     """For 3-4-p sphere model."""
#     return 1.5*q**2 + 2*q**3
```

```python
step_exp = 10
n_tot = 2**step_exp
h = 0.01
beta0 = 0
beta = 2
Tf = 0.5
```

```python
C = np.nan*np.ones((n_tot, n_tot))
R = np.nan*np.ones((n_tot, n_tot))
C[0][0] = R[0][0] = 1
u = [Tf + beta0 * df1(C[0][0])*C[0][0]]
E = [-beta0*f(C[0][0])]
```

```python
fig = plt.figure()
ax = fig.add_subplot(1,1,1)
plt.ion()
plt.grid()
plt.xscale('log')
plt.show()
```

```python
for t_i in range(0, n_tot):
    for t1_i in range(t_i + 1):
        int11 = 0
        for s_i in range(t_i):
            height = df2(C[s_i][t_i])*R[s_i][t_i]*C[t1_i][s_i]
            width = h
            area = height * width
            int11 += area
        
        int12 = 0
        for s_i in range(t1_i):
            height = df1(C[s_i][t_i])*R[s_i][t1_i]
            width = h 
            area = height * width
            int12 += area
        
        C[t_i+1][t1_i] = C[t1_i][t_i+1] = C[t1_i][t_i] + h*(-1*u[t_i]*C[t1_i][t_i] + \
                                    beta0*df1(C[0][t_i])*C[0][t1_i] + \
                                          int11 + int12) 
        int21 = 0 
        for s_i in range(t1_i, t_i):
            height = df2(C[s_i][t_i])*R[s_i][t_i]*R[t1_i][s_i]
            width = h 
            area = height * width
            int21 += area

        R[t1_i][t_i+1] = R[t1_i][t_i] +h*(-1*u[t_i]*R[t1_i][t_i] +\
                                           int21) 
    C[t_i+1][t_i+1] = 1
    R[t_i+1][t_i+1] = 1
    print( t_i + 1, ' done')

    int31 = 0
    for s_i in range(0, t_i+1):
        height = df2(C[s_i][t_i+1])*R[s_i][t_i+1]*C[s_i][t_i+1]
        width = h 
        area = height * width
        int31 += area

    int32 = 0
    for s_i in range(0, t_i+1):
        height = df1(C[s_i][t_i+1])*R[s_i][t_i+1]
        width = h 
        area = height * width
        int32 += area
    u.append(Tf + beta0*df1(C[0][t_i+1])*C[0][t_i+1] + int31 + int32)
    print('Debug u:', u)

    int41 = 0 
    for s_i in range(0, t_i+1):
        height = df1(C[s_i][t_i+1])*R[s_i][t_i+1]
        width = h
        area = height * width
        int41 += area
    E.append(-1*int41 - beta0*f(C[0][t_i+1]))
    print('Debug E', E)
    plt.plot(E)
    plt.pause(0.1)
```
